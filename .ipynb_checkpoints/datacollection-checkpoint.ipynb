```python
# Importation des packages
import pandas as pd
from requests import get
from bs4 import BeautifulSoup as bs
```

```python
url1 = 'https://sn.coinafrique.com/categorie/chiens?page=1' #V1: Nom, V2 : prix, V3 : adresse, V4: image_lien
res = get(url1)
soup = bs(res.content, 'html.parser')
# print(soup)
containers = soup.find_all('div','col s6 m4 l3')
len(containers)

# url2 = https://sn.coinafrique.com/categorie/moutons?page=1 #V1: Nom, V2 : prix, V3 : adresse, V4: image_ln

# url3 = https://sn.coinafrique.com/categorie/poules-lapins-et-pigeons?page=1 #V1: details, V2 : prix, V3 : adresse, V4: image_lien

# url4 = https://sn.coinafrique.com/categorie/autres-animaux?page=1 #V1: Nom, V2 : prix, V3 : adresse, V4: image_lien
```
