```python
container = containers[0]

#V1: Nom
nom = container.find('p','ad__card-description').a.text
#V2: prix
prix = container.find('p','ad__card-price').a.text
#V3: adresse
adresse = container.find('p','ad__card-location').span.text
#V4: image_lien
image_lien = container.find('img','ad__card-img')['src']
```

```python
df_chien = pd.DataFrame()
for index_page in range(1, 5):
    url1 = f'https://sn.coinafrique.com/categorie/chiens?page={index_page}' #V1: Nom, V2 : prix, V3 : adresse, V4: image_lien
    res = get(url1)
    soup = bs(res.content, 'html.parser')
    # print(soup)
    containers = soup.find_all('div','col s6 m4 l3')
    data = []
    for container in containers:
        try:
            nom = container.find('p','ad__card-description').a.text
            prix = container.find('p','ad__card-price').a.text
            adresse = container.find('p','ad__card-location').span.text
            image_lien = container.find('img','ad__card-img')['src']
            dic = {
                'nom': nom,
                'prix': prix,
                'adresse': adresse,
                'image_lien': image_lien
            }
            data.append(dic)
        except:
            pass
            
    df = pd.DataFrame(data)
    df_chien = pd.concat([df_chien, df], axis = 0).reset_index(drop = True)

```
